```python
# Import necessary libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display

# Sample Arabic text (you can replace this with your own text)

# get the data from file.txt
with open('file.txt', 'r', encoding='utf-8') as file:
    arabic_text = file.read()
    

# Reshape the Arabic text for correct display
reshaped_text = arabic_reshaper.reshape(arabic_text)

# Get the correct display order (right-to-left for Arabic)
bidi_text = get_display(reshaped_text)

# Generate the word cloud
wordcloud = WordCloud(font_path='arial', background_color='white', width=800, height=600).generate(bidi_text)

# Display the generated word cloud
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
